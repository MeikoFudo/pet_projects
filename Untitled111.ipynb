```python
import os
import requests
from bs4 import BeautifulSoup

# Путь к файлу с ссылками
filename = "/content/clean_links_news.txt"

# Создание папки для сохранения файлов, если она не существует
output_folder = "TEXTSPARSE1"
os.makedirs(output_folder, exist_ok=True)

# Открытие файла с ссылками
with open(filename, "r") as file:
    links = file.readlines()

# Парсинг текста со страниц
parsed_count = 0  # Счетчик спарсенных текстов

for link in links:
    link = link.strip()  # Удаление лишних пробелов и символов новой строки

    # Отправка запроса и получение содержимого страницы
    response = requests.get(link)
    soup = BeautifulSoup(response.content, "html.parser")

    # Поиск элементов с заданными атрибутами
    headers = soup.find_all(class_="single-header")
    texts = soup.find_all(style="text-align: justify;")
    div_texts = soup.find_all("div", {"align": "justify"})
    p_texts = soup.find_all("p")

    # Объединение всех найденных текстовых элементов
    all_texts = []
    if headers:
        all_texts.append(headers[0].text.strip())
    all_texts += [text.text.strip() for text in texts]
    all_texts += [div.text.strip() for div in div_texts]
    all_texts += [p.text.strip() for p in p_texts]

    # Если найден хотя бы один текст
    if all_texts:
        parsed_count += 1
        filename = f"parsed_text_{parsed_count}.txt"
        filepath = os.path.join(output_folder, filename)

        with open(filepath, "w") as output_file:
            output_file.write("\n".join(all_texts))

        print(f"Успешно спарсен текст со ссылки {link}.")
        print("Сохранен в файле:", filepath)
        print("Текст:\n", "\n".join(all_texts))
        print("-------------------------------------------")
    else:
        print("Не удалось спарсить текст по ссылке:", link)
        print("-------------------------------------------")

# Вывод общего количества спарсенных текстов
print("Всего спарсено текстов:", parsed_count)
```
